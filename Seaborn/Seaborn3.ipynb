# KDE plot

**kernel density estimation (KDE)** is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

```python
mean = [0,0]
cov = [[0.2,0],[0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean,cov,size = 40).T
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis)
```

```python
sns.kdeplot(x_axis,shade = True,color = "orange")
```

```python
sns.kdeplot(x_axis,y_axis)
```

```python
sns.kdeplot(x_axis,y_axis,shade = True,color = "g")
```

```python
sns.kdeplot(x_axis,y_axis,n_levels = 10,cmap = "Blues")
```

```python
sns.kdeplot(y_axis,bw = 0.4,color = "g",shade = True)
```

```python
iris = sns.load_dataset("iris")
iris.head(4)
```

```python
setosa = iris.loc[iris.species == "setosa"]
versicolor = iris.loc[iris.species == "versicolor"]

sns.kdeplot(setosa.petal_length,setosa.petal_width)
sns.kdeplot(versicolor.sepal_length,versicolor.sepal_width)
```

```python
sns.kdeplot(setosa.petal_length,setosa.petal_width, cmap = "RdBu")
sns.kdeplot(versicolor.sepal_length,versicolor.sepal_width , cmap = "coolwarm",shade = True)
```

```python
sns.kdeplot(versicolor.petal_width,versicolor.petal_length,n_labels = 30,cmap = "gnuplot")
```

# Joint Plot

```python
iris = sns.load_dataset("iris")

iris.head()
```

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris)
```

```python
sns.jointplot(x = "petal_width", y = "petal_width", data = iris, kind = "hex",cmap = "cool")
```

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, kind = "reg", color ="g")
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
sns.jointplot(x = "total_bill", y = "tip",data = tips)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "reg", color = "b")
```

```python
sns.jointplot(x = "size", y = "total_bill", data = tips,kind = "kde")
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "reg", color = "b",height = 5, ratio = 2)
```

# Regression Plot (Reg plot)

```python
sns.regplot(x = "sepal_length", y = "sepal_width",data =iris,color = "k")
```

```python
sns.regplot( x= "total_bill", y = "tip", data = tips,color = "purple")
```

```python
mean = [2,5]
cov = [[3.3,4], [4.3,3]]

x_val, y_val = np.random.multivariate_normal(mean, cov,100).T
```

```python
sns.regplot(x = x_val, y = y_val, color = "blue")
```

```python
sns.regplot(x = "total_bill",y = "tip", data = tips,marker = "D",
            scatter_kws = {"color":"yellow"},
            line_kws = {"color":"k","linewidth":3.2})
```

```python
sns.regplot(x = "size", y = "total_bill",data = tips,color = "r")
```

```python
sns.regplot(x = "size", y ="total_bill", data = tips,x_jitter =0.08,color = "g")
```

```python
sns.regplot(x = "size", y ="total_bill", data = tips,x_estimator = np.mean,color = "g")
```

# Pair Plot

To plot multiple pairwise bivariate distributions in a dataset, we can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue = "species",palette = "Blues_d",markers = ["*","o","D"])
```

```python
sns.pairplot(iris, vars = ["sepal_length","sepal_width"],palette = "husl",hue = "species")
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = "smoker", palette = "inferno")
```

```python
sns.pairplot(tips,diag_kind = "kde", kind = "reg",hue = "sex",palette = "Blues_d")
```
