```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Violin Plot

```python
tips = sns.load_dataset('tips')
tips.head(4)
```

```python
sns.violinplot(x = "size",y= "total_bill",data = tips)
```

```python
sns.violinplot(x= "day",y = "total_bill",hue = "sex",data = tips)
```

```python
sns.violinplot(x = "sex",y = "tip",hue ="smoker",palette = "inferno",data = tips)
```

```python
sns.violinplot(x = "smoker",y = "tip",hue = "sex",split = True,data = tips)
```

```python
sns.violinplot(x = "sex",y = "tip",hue ="smoker",palette = "magma_r",data = tips,
               inner = "quartile")
```

```python
sns.violinplot(x= "day",y = "total_bill",hue = "sex",data = tips,
               inner = "quartile",scale = "count")
```

```python
sns.violinplot(x= "day",y = "total_bill",hue = "sex",data = tips,
               inner = "stick",scale = "count")
```

```python
sns.violinplot(x = "day",y = "tip",data = tips,order = ["Sat","Sun","Thur","Fri"],
               hue = "sex",palette = "mako",
               inner = "stick",scale = "count",scale_hue = False,bw = 0.7)
```

# ClusterMap

```python
flight = sns.load_dataset("flights")
flight.head(5)
```

```python
flight = flight.pivot("month","year","passengers")
sns.clustermap(flight)
```

```python
sns.clustermap(flight,col_cluster = False,cmap = "coolwarm")
```

```python
sns.clustermap(flight,row_cluster = False,linewidth = 1.4,cmap = "coolwarm_r")
```

```python
# Standardize across col and data

sns.clustermap(flight,standard_scale = 1)
```

```python
sns.clustermap (flight,standard_scale = 0.8)
```

```python
sns.clustermap(flight,z_score =1)   ## z_score = 0 is row
```

```python
sns.clustermap(flight,z_score = 0)
```

# Heatmap

**Heatmap** is defined as a graphical representation of data using colors to visualize the value of the matrix. In this, to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred. Heatmap is also defined by the name of the shading matrix. Heatmaps in Seaborn can be plotted by using the **seaborn.heatmap()** function.



```python
normal = np.random.rand(10,10)

#heatmap
plt.figure(figsize = (6,6))
sns.heatmap(normal,annot = True)
```

```python
flight = sns.load_dataset("flights")
flight = flight.pivot("month","year","passengers")
flight.head(5)
```

```python
sns.heatmap(flight,annot = True,vmin = 100,vmax = 550,fmt = "d")
```

```python
sns.heatmap(flight,center = flight.loc["August",1954],cmap = "coolwarm",
            annot = True,fmt= "d",linewidth = 0.6)
```

```python
flight.corr()
```

```python
plt.figure(figsize = (7,7))

sns.heatmap(flight.corr(),annot = True,fmt = "d",
            cmap = "winter_r",linewidth = 0.8)

plt.title("Correlation Plot",fontweight = "bold",fontsize = 18)
```

# FacetGrid 

* FacetGrid class helps in visualizing distribution of one variable as well as the relationship between multiple variables   separately within subsets of your dataset using multiple panels..

* A FacetGrid can be drawn with up to three dimensions . row, col, and hue. The first two have obvious correspondence with the resulting array of axes; think of the hue variable as a third dimension along a depth axis, where different levels are plotted with different colors.

```python
plt.style.use("ggplot")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.FacetGrid(tips,row = "time",col = "smoker")
```

```python
x = sns.FacetGrid(tips,row = "smoker",col = "time")
x = x.map(plt.hist,"total_bill",color = "green")
```

```python
x = sns.FacetGrid(tips,row = "time",col = "smoker")
x = x.map(plt.hist,"tip",color = "blue")
```

```python
x = sns.FacetGrid(tips,row = "time",col = "smoker")
x = x.map(plt.scatter,"total_bill","tip",color = "k")
```

```python
x = sns.FacetGrid(tips,row = "time",col = "smoker")
x = x.map(sns.regplot,"total_bill","tip",color = "red")
```

```python
x = sns.FacetGrid(tips, col = "day")
x = x.map(sns.boxplot,"total_bill","time")
```

```python
# adding legend

x  = sns.FacetGrid(tips,col = "time",hue = "sex")
x = (x.map(plt.scatter,"total_bill","tip").add_legend())
```

```python
x = sns.FacetGrid(tips,col = "day",size = 7,aspect = .4)
x = x.map(sns.boxplot,"time","total_bill")
```
