# Box plot

Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used for detect the outlier in data set. It captures the summary of the data efficiently with a simple box and whiskers and allows us to compare easily across groups. Boxplot summarizes a sample data using 25th, 50th and 75th percentiles. These percentiles are also known as the lower quartile, median and upper quartile.

```python
tips = sns.load_dataset("tips")
tips.head(4)
```

```python
sns.boxplot(x = tips["size"])
```

```python
sns.boxplot(tips["total_bill"])
```

```python
sns.boxplot(x = "sex",y = "total_bill",data = tips)
```

```python
sns.boxplot(x = "day",y = "total_bill",hue = "sex",data = tips)
```

```python
sns.boxplot(x = "day",y = "tip",hue = "smoker",data = tips,palette = "bright")
```

```python
sns.boxplot(x = "day",y = "total_bill",order = ["Sat","Sun","Thur","Fri"],data = tips,palette = "bright")
```

```python
sns.boxplot(x = "day", y = "tip",data = tips,color = "blue")
sns.swarmplot(x = "day", y = "tip",data = tips,color = "black")
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
ax = sns.boxplot(x="day", y="total_bill", hue="weekend",data=tips, dodge=True)
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
ax = sns.boxplot(x="day", y="total_bill", hue="weekend",data=tips, dodge=False)
```

```python
sns.catplot(x = "day", y = "total_bill",data = tips)
```

```python
df = pd.read_csv("heart-disease.csv")

sns.boxplot(x = "",y = "target",data = df)
```

# Strip Plot

A strip plot is drawn on its own. It is a good complement to a boxplot or violinplot in cases where all observations are shown along with some representation of the underlying distribution. It is used to draw a scatter plot based on the category.

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

```python
sns.stripplot(tips["tip"],color = "green")
```

```python
sns.stripplot(x= "day",y = "tip",data = tips)
```

```python
sns.stripplot(x = "day",y = "total_bill",data = tips,jitter = False)
```

```python
sns.stripplot(x = "total_bill",y = "day",hue = "sex",data = tips,palette = "BuPu")
```

```python
sns.stripplot(x = "day",y = "tip",data = tips,hue = "smoker",jitter = 0.2,linewidth = 1.3,palette = "Wistia")
```

```python
sns.stripplot(x = "day", y = "tip",data= tips,hue = "sex",dodge = True,color = "yellow")
```

```python
sns.stripplot(x = "day",y = "total_bill",data = tips,order = ["Sat","Sun","Thur","Fri"],hue = "smoker",
              dodge = True,color = "cyan")
```

```python
sns.stripplot(x = "sex",y= "tip",order = ["Female","Male"],hue = "smoker",data = tips,marker = "D",color = "blue",dodge = True)
```

```python
sns.stripplot(x = "day",y = "total_bill",data = tips,color = "k",marker = "D",edgecolor = "gray")
sns.boxplot(x = "day", y = "total_bill",data = tips)
```

# Pair Grid

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.PairGrid(iris)
```

```python
x = sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris,hue = "species",palette = "cool")
x = x.map_diag(plt.hist,histtype = "step",linewidth = 3)
x = x.map_offdiag(plt.scatter)
plt.legend()
```

```python
g = sns.PairGrid(iris)
g = g.map_upper(plt.scatter)
g = g.map_lower(sns.kdeplot,cmap = "Blues_r")
g = g.map_diag(sns.kdeplot,lw = 3,legend = False)
plt.show()
```

```python
x = sns.PairGrid(iris,vars = ['petal_length','petal_width'],hue = "species",hue_kws = {"marker":["D","s","+"]})
x = x.map_diag(plt.hist,edgecolor = "k")
x = x.map_offdiag(plt.scatter,edgecolor = "k")
```

```python
x = sns.PairGrid(iris,x_vars = ["sepal_length","sepal_width"],
                 y_vars = ["petal_length","petal_width"])
x = x.map(plt.scatter,edgecolor = "r")
```
