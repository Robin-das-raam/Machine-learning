# Bar chart

```python
x1 = np.arange(25,30)
y1 = [12000,14000,16000,15000,18000]
plt.bar(x1,y1,color = "#abbcdc",linestyle = "--",label = "mechanic") 

x2 = np.arange(25,30)
y2 = [25000,27000,28000,26000,23000]
plt.plot(x2,y2,color = "#454564",marker = "o",label = "engineer")

y3 = [20000,21600,23000,22000,24300]
plt.plot(x1,y3,color = "#787865",marker = "*",label = "Doctor")

plt.title("age vs salary")
plt.xlabel("age")
plt.ylabel("salary")
plt.legend()  
plt.show()
```

Above figure shows a bar chart with two line plot. 

```python
x = np.arange(25,30)
x1 = np.arange(len(x))
y1 = [12000,14000,16000,15000,18000]
plt.bar(x1,y1,color = "#abbcdc",linestyle = "--",label = "mechanic") 


y2 = [25000,27000,28000,26000,23000]
plt.bar(x1,y2,color = "#454564",label = "engineer")

y3 = [20000,21600,23000,22000,24300]
plt.bar(x1,y3,color = "#787865",label = "Doctor")

plt.title("age vs salary")
plt.xlabel("age")
plt.ylabel("salary")
plt.legend()  
plt.show()
```

This is not an appropiate way to plot a bar chart.Lets do it on right way

```python
x = np.arange(25,30)
x1 = np.arange(len(x))
width = 0.25

y1 = [12000,14000,16000,15000,18000]
plt.bar(x1-width,y1,width = width,color = "#abbcdc",linestyle = "--",label = "mechanic") 


y2 = [25000,27000,28000,26000,23000]
plt.bar(x1,y2,width = width,color = "#454564",label = "engineer")

y3 = [20000,21600,23000,22000,24300]
plt.bar(x1+width,y3,width = width,color = "#787865",label = "Doctor")

plt.title("age vs salary")
plt.xticks(ticks = x1,labels = x)
plt.xlabel("age")
plt.ylabel("salary")
plt.legend()  
plt.show()
```

# Stack Plot

**Stack Plots are generated by plotting different datasets vertically on top of one another rather than overlapping with one another. Matplotlib has a defined function for creating stack plot matplotlib.pyplot.stackplot(). It is often used to visualize data in the form of summation.Stack Plots are used to plot linear data, in a vertical order, stacking each linear plot on another. Typically, they're used to generate cumulative plots.**

```python
minutes = [1,2,3,4,5,6,7,8,9]

player_1 = [1,2,2,3,1,3,4,4,5]
player_2 = [1,1,1,2,3,3,4,5,3]
player_3 = [1,2,2,3,3,4,4,4,2]

plt.stackplot(minutes,player_1,player_2,player_3)
plt.title("Simple stack plot")
plt.tight_layout()
plt.show()
```

```python
minutes = [1,2,3,4,5,6,7,8,9]

player_1 = [1,2,2,3,1,3,4,4,5]
player_2 = [1,1,1,2,3,3,4,5,3]
player_3 = [1,2,2,3,3,4,4,4,2]

labels = ["player_1","player_2","player_3"]

plt.stackplot(minutes,player_1,player_2,player_3,labels = labels)
plt.title("Simple stack plot")
plt.tight_layout()
#plt.legend()
#defining the exact location for legend
plt.legend(loc = "upper left")
plt.show()
```

```python
days = np.arange(1,10)

dev_1 = [8,6,5,4,3,5,3,5,1]
dev_2 = [0,1,1,2,2,1,4,0,5]
dev_3 = [0,1,2,2,3,2,1,3,2]

labels = ["dev_1","dev_2","dev_3"]
color = ["c","m","k"]

plt.stackplot(days,dev_1,dev_2,dev_3,labels = labels,colors = color)
plt.title("Stack plot")
#plt.legend(loc = "lower left")
plt.legend(loc = (0.07,0.05))
plt.show()
```

# Histogram

* A histogram is an accurate graphical representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable (quantitative variable) and was first introduced by Karl Pearson. It is a type of bar graph. To construct a histogram, the first step is to “bin” the range of values — that is, divide the entire range of values into a series of intervals — and then count how many values fall into each interval. The bins are usually specified as consecutive, non-overlapping intervals of a variable. The bins (intervals) must be adjacent and are often (but are not required to be) of equal size.

```python
ages = [17,18,20,22,23,24,26,30,34,36,38,40,45,47,50]

plt.hist(ages,bins = 5,edgecolor = "black")
plt.title("Age")
plt.show()
```

```python
ages = [17,18,20,22,23,24,26,30,34,36,38,40,45,47,50]
bins = np.arange(10,60,10)

plt.hist(ages,bins = bins,edgecolor = "black")
plt.title("Age")
plt.show()
```

```python
mu,sigma = 100,20
x = mu +sigma * np.random.randn(10000)

bins = 50

plt.hist(x,bins= bins,density = True,alpha = 0.75)
plt.xlabel('Values')
plt.ylabel('Probability Distribution')
plt.title('Histogram showing Data Distribution')
plt.text(60, .025, r'$\mu=100,\ \sigma=20$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```
